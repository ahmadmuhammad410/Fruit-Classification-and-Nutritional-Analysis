```python
import os
import time
import random
import numpy as np
import cv2
import matplotlib.pyplot as plt
from pathlib import Path
from tqdm.auto import tqdm

# --- 1. Environment Setup & Dependency Resolution ---
# Force specific versions to fix NumPy 2.0 and Transformers conflicts
print("Initializing environment...")
os.system('pip install -q "numpy<2.0.0" "transformers>=4.40.0" segmentation-models-pytorch albumentations torchinfo torchmetrics')

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

# Robust Mixed Precision Imports
try:
    from torch.amp import GradScaler, autocast
    USE_NEW_AMP = True
except ImportError:
    from torch.cuda.amp import GradScaler, autocast
    USE_NEW_AMP = False

import segmentation_models_pytorch as smp
# Fix for the ImportError: Importing from the correct sub-modules
from torchmetrics.classification import (
    BinaryJaccardIndex, 
    BinaryF1Score, 
    BinaryPrecision, 
    BinaryRecall, 
    BinaryAccuracy
)

# --- 2. Configuration & Seeds ---
class CFG:
    seed = 42
    base_path = "/kaggle/input/cv-dataset/Project Data/Fruit"
    ckpt_dir = "/kaggle/working/checkpoints"
    os.makedirs(ckpt_dir, exist_ok=True)
    
    # Model Architecture
    model_arch = 'UnetPlusPlus'
    backbone = 'efficientnet-b3'
    img_size = (256, 256)
    
    # Training Params
    epochs = 50
    batch_size = 16
    lr = 1e-3
    min_lr = 1e-6
    weight_decay = 1e-4
    grad_accum = 2
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    
    # Resume / Save paths
    resume = True
    latest_ckpt = f"{ckpt_dir}/latest_checkpoint.pth"
    best_ckpt = f"{ckpt_dir}/best_binary_model.pth"

def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

seed_everything(CFG.seed)

# --- 3. Dataset Class with Sanity Checks ---
class FruitBinaryDataset(Dataset):
    def __init__(self, root_dir, mode='Train'):
        self.root_dir = Path(root_dir) / mode
        self.images = []
        self.masks = []
        
        if not self.root_dir.exists():
            raise FileNotFoundError(f"Path not found: {self.root_dir}")

        # Crawl categories (Apple, Banana, etc.)
        categories = sorted([d for d in self.root_dir.iterdir() if d.is_dir()])
        for cat in categories:
            img_folder = cat / "Images"
            mask_folder = cat / "Mask"
            
            img_paths = sorted(list(img_folder.glob("*.jpg")))
            for p in img_paths:
                m = mask_folder / f"{p.stem}_mask.png"
                # REQUIREMENT: Fail loudly if mask is missing
                if not m.exists():
                    raise FileNotFoundError(f"CRITICAL ERROR: Mask missing for image {p}")
                
                self.images.append(str(p))
                self.masks.append(str(m))
        
        print(f"[{mode}] Loaded {len(self.images)} samples.")

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        # Load
        img = cv2.imread(self.images[idx])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        mask = cv2.imread(self.masks[idx], cv2.IMREAD_GRAYSCALE)
        
        # Resize
        img = cv2.resize(img, CFG.img_size)
        mask = cv2.resize(mask, CFG.img_size, interpolation=cv2.INTER_NEAREST)
        
        # Binarize mask
        mask = (mask > 127).astype(np.float32)
        
        # To Tensor
        img = torch.from_numpy(img).permute(2, 0, 1).float() / 255.0
        mask = torch.from_numpy(mask).unsqueeze(0).float()
        
        return img, mask

# --- 4. Training Utilities ---
class HybridLoss(nn.Module):
    """Wrapper to combine Dice and Focal loss since smp might not support + operator"""
    def __init__(self):
        super().__init__()
        self.dice = smp.losses.DiceLoss(mode='binary')
        self.focal = smp.losses.FocalLoss(mode='binary')
        
    def forward(self, logits, targets):
        return self.dice(logits, targets) + self.focal(logits, targets)

def build_model():
    model = smp.UnetPlusPlus(
        encoder_name=CFG.backbone,
        encoder_weights="imagenet",
        in_channels=3,
        classes=1,
        activation=None # Use raw logits for better stability with Focal Loss
    )
    return model

def visualize_sample(img, mask, pred, epoch):
    img = img[0].cpu().permute(1, 2, 0).numpy()
    mask = mask[0].cpu().squeeze().numpy()
    pred = torch.sigmoid(pred[0]).cpu().detach().squeeze().numpy() > 0.5
    
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 3, 1); plt.imshow(img); plt.title("Input")
    plt.subplot(1, 3, 2); plt.imshow(mask, cmap='gray'); plt.title("Ground Truth")
    plt.subplot(1, 3, 3); plt.imshow(pred, cmap='gray'); plt.title(f"Pred Epoch {epoch}")
    plt.show()

# --- 5. Main Execution Block ---
def main():
    # Load Data
    train_ds = FruitBinaryDataset(CFG.base_path, 'Train')
    val_ds = FruitBinaryDataset(CFG.base_path, 'Validation')
    
    train_loader = DataLoader(train_ds, batch_size=CFG.batch_size, shuffle=True, num_workers=2)
    val_loader = DataLoader(val_ds, batch_size=CFG.batch_size, shuffle=False, num_workers=2)
    
    # Initialize Model, Optimizer, Loss
    model = build_model().to(CFG.device)
    optimizer = optim.AdamW(model.parameters(), lr=CFG.lr, weight_decay=CFG.weight_decay)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=CFG.epochs, eta_min=CFG.min_lr)
    
    # Combined Loss: Dice + Focal for robust segmentation
    criterion = HybridLoss()
    
    # Updated GradScaler for modern torch.amp
    if USE_NEW_AMP:
        scaler = GradScaler('cuda' if torch.cuda.is_available() else 'cpu')
    else:
        scaler = GradScaler()
    
    # Metrics (F1Score is Dice for binary)
    metrics = {
        "iou": BinaryJaccardIndex().to(CFG.device),
        "dice": BinaryF1Score().to(CFG.device),
        "acc": BinaryAccuracy().to(CFG.device),
        "prec": BinaryPrecision().to(CFG.device),
        "rec": BinaryRecall().to(CFG.device)
    }
    
    start_epoch = 0
    best_iou = 0.0
    
    # RESUME LOGIC
    if CFG.resume and os.path.exists(CFG.latest_ckpt):
        print(f"Loading checkpoint: {CFG.latest_ckpt}")
        ckpt = torch.load(CFG.latest_ckpt, map_location=CFG.device)
        model.load_state_dict(ckpt['model_state_dict'])
        optimizer.load_state_dict(ckpt['optimizer_state_dict'])
        scheduler.load_state_dict(ckpt['scheduler_state_dict'])
        start_epoch = ckpt['epoch'] + 1
        best_iou = ckpt['best_iou']
        print(f"Resumed from epoch {start_epoch}")

    # --- Training Loop ---
    for epoch in range(start_epoch, CFG.epochs):
        model.train()
        train_loss = 0
        pbar = tqdm(train_loader, desc=f"Epoch {epoch}/{CFG.epochs}")
        
        for i, (imgs, masks) in enumerate(pbar):
            imgs, masks = imgs.to(CFG.device), masks.to(CFG.device)
            
            # Use appropriate autocast syntax based on version
            autocast_kwargs = {"device_type": 'cuda'} if USE_NEW_AMP and torch.cuda.is_available() else {}
            
            with autocast(**autocast_kwargs):
                logits = model(imgs)
                loss = criterion(logits, masks)
                loss = loss / CFG.grad_accum
            
            scaler.scale(loss).backward()
            
            if (i + 1) % CFG.grad_accum == 0:
                scaler.step(optimizer)
                scaler.update()
                optimizer.zero_grad()
            
            train_loss += loss.item() * CFG.grad_accum
            pbar.set_postfix({'loss': train_loss / (i+1)})
            
        # --- Validation ---
        model.eval()
        val_loss = 0
        for m in metrics.values(): m.reset()
        
        start_latency = time.time()
        with torch.no_grad():
            for imgs, masks in val_loader:
                imgs, masks = imgs.to(CFG.device), masks.to(CFG.device)
                logits = model(imgs)
                loss = criterion(logits, masks)
                val_loss += loss.item()
                
                preds = (torch.sigmoid(logits) > 0.5).float()
                for m in metrics.values(): m.update(preds, masks)
        
        latency = (time.time() - start_latency) / len(val_ds) * 1000 # ms per image
        
        # Calculate Epoch Results
        epoch_iou = metrics['iou'].compute().item()
        epoch_dice = metrics['dice'].compute().item()
        
        print(f"\n[Epoch {epoch}] Val Loss: {val_loss/len(val_loader):.4f} | mIoU: {epoch_iou:.4f} | Dice: {epoch_dice:.4f}")
        print(f"Inference Latency: {latency:.2f}ms/image")
        
        # REQUIREMENT: Periodic visualization
        if epoch % 5 == 0:
            visualize_sample(imgs, masks, logits, epoch)
            
        # Save Checkpoints
        ckpt_data = {
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'scheduler_state_dict': scheduler.state_dict(),
            'best_iou': best_iou
        }
        torch.save(ckpt_data, CFG.latest_ckpt)
        
        if epoch_iou > best_iou:
            best_iou = epoch_iou
            torch.save(model.state_dict(), CFG.best_ckpt)
            print(">> Best model saved!")
            
        scheduler.step()

if __name__ == "__main__":
    main()
```
